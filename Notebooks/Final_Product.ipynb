## Part 2: Description of Final Product

### Overview
This infographic visualizes the distribution and characteristics of public art across Virginia using data from the Public Art Archive. The visualization explores where public art is concentrated geographically, when artworks were created, what types are most common, and how Virginia's public art landscape compares to global trends.

### Reflection
The infographic reveals several key patterns in Virginia's public art. Northern Virginia's dominance (over half of all documented public art) shows that proximity to Washington D.C. and population density drive public art investment. This concentrates most of Virginia's public art in one region, which may raise questions about access in rural and less populated areas.

Virginia also favors public spaces and park grounds over government office buildings, which differs from global trends. This may suggest that Virginia communities prioritize accessible outdoor art rather than art in formal institutions.

The sharp drop in completed artworks around 2022 aligns with COVID-19 and could link to the disruption of funding and resources at the time. This shows how vulnerable public art is to economic crisis.

Overall, these patterns show that Virginia's public art landscape is shaped by geography, community values, and economic conditions. These factors matter and may be useful observations that could be used in future public art policy and funding decisions.

### Design Choices
I wanted the infographic to use the Public Art Archive's official brand colors, fonts, and logos from their [media kit](https://s3.amazonaws.com/marketing.westaf.org/PAA+Marketing/FY25+Public+Art+Archive+Media+Kit.pdf) to maintain visual consistency with the organization. I additionally wanted to draw attention to key points on graphs with annotations that guide the viewer through the key findings. Lastly, I wanted to reduce the busyness of texts by adding icons that relate to the theme.

My goal was to implement class concepts by answering these questions:
- Motivation - Why?
  - To highlight how Virginia's public art ecosystem differs from global patterns and reveal regional concentrations
- Form - How?
  - Through comparative bar charts, time-series visualization, and annotated geographic mapping
- Content - What?
  - Public art placement types, artwork creation timelines, and geographic distribution data
- Audience - Who?
  - Researchers, artists, community members, and general public interested in Virginia's public art landscape
- Situation - Where? When?
  - Virginia's public art collection as documented in the Public Art Archive (as of December 2025)

I kept these answers in mind during the pre-processing and post-processing phases to ensure a productive output of explanatory analysis.

### Data Source
All data was collected from the Public Art Archive public API (https://publicartarchive.org/), filtered to create two datasets: one global and one merged and filtered to include only artworks located in Virginia, which includes 319 individual artworks. My process of acquiring and cleaning this data can be found [here.](https://github.com/kenzkleczynski/Public-Art-Archive/blob/main/Notebooks/Establishing_Data.ipynb)

### Limitations
- The Public Art Archive is not a complete census as it only includes artworks that have been submitted to the database.
- There may be collection bias toward institutional art, with grassroots public art potentially underrepresented due to lack of organizations or individuals uploading this work to the archive.
- Data reflects the archive as of December 2025.
